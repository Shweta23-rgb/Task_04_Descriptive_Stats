```python
import time
import math
from collections import Counter

# Script 1 Part 1: Pure Python Descriptive Statistics for Facebook Dataset
print("=== SCRIPT 1 PART 1: PURE PYTHON OVERALL STATISTICS ===")

# Start timing
start_time = time.time()

# Convert DataFrame to list of dictionaries for pure Python processing
facebook_data = facebook_posts_clean.to_dict('records')

# Define our column categories (same as Pandas)
numeric_columns = [
    'Likes', 'Comments', 'Shares', 'Love', 'Wow', 'Haha', 'Sad', 'Angry', 'Care',
    'Post Views', 'Total Views', 'Total Views For All Crossposts', 'Overperforming Score',
    'Total Interactions',
    'advocacy_msg_type_illuminating', 'issue_msg_type_illuminating', 'attack_msg_type_illuminating',
    'image_msg_type_illuminating', 'cta_msg_type_illuminating', 'engagement_cta_subtype_illuminating',
    'fundraising_cta_subtype_illuminating', 'voting_cta_subtype_illuminating',
    'covid_topic_illuminating', 'economy_topic_illuminating', 'education_topic_illuminating',
    'environment_topic_illuminating', 'foreign_policy_topic_illuminating', 'governance_topic_illuminating',
    'health_topic_illuminating', 'immigration_topic_illuminating', 'lgbtq_issues_topic_illuminating',
    'military_topic_illuminating', 'race_and_ethnicity_topic_illuminating', 'safety_topic_illuminating',
    'social_and_cultural_topic_illuminating', 'technology_and_privacy_topic_illuminating',
    'womens_issue_topic_illuminating', 'incivility_illuminating', 'scam_illuminating',
    'freefair_illuminating', 'fraud_illuminating'
]

categorical_columns = [
    'Facebook_Id', 'post_id', 'Page Category', 'Page Admin Top Country', 
    'Post Created', 'Type', 'Video Share Status', 'Is Video Owner?', 'Video Length'
]

print("OVERALL DATASET STATISTICS")
print("=" * 60)

print(f"\nDataset Info:")
print(f"Total Rows: {len(facebook_data)}")
print(f"Total Columns: {len(facebook_data[0].keys()) if facebook_data else 0}")
print(f"Numeric Columns: {len(numeric_columns)}")
print(f"Categorical Columns: {len(categorical_columns)}")

# HELPER FUNCTIONS FOR PURE PYTHON STATISTICS
def get_numeric_values(data, column):
    """Extract numeric values from a column, handling NaN/None"""
    values = []
    for row in data:
        val = row.get(column)
        if val is not None and not (isinstance(val, float) and math.isnan(val)):
            try:
                values.append(float(val))
            except (ValueError, TypeError):
                pass
    return values

def calculate_mean(values):
    """Calculate mean"""
    return sum(values) / len(values) if values else 0

def calculate_std(values):
    """Calculate standard deviation"""
    if len(values) < 2:
        return 0
    mean = calculate_mean(values)
    variance = sum((x - mean) ** 2 for x in values) / len(values)
    return math.sqrt(variance)

def calculate_percentile(sorted_values, percentile):
    """Calculate percentile of sorted values"""
    if not sorted_values:
        return 0
    k = (len(sorted_values) - 1) * percentile / 100
    f = math.floor(k)
    c = math.ceil(k)
    if f == c:
        return sorted_values[int(k)]
    return sorted_values[int(f)] * (c - k) + sorted_values[int(c)] * (k - f)

# NUMERIC STATISTICS
print("\n" + "=" * 60)
print("NUMERIC COLUMNS STATISTICS")
print("=" * 60)

print("\nPure Python numeric statistics (equivalent to .describe()):")
print(f"{'Column':<45} {'count':<8} {'mean':<10} {'std':<10} {'min':<10} {'25%':<10} {'50%':<10} {'75%':<10} {'max':<12}")
print("-" * 150)

numeric_stats_python = {}
for col in numeric_columns:
    values = get_numeric_values(facebook_data, col)
    
    if values:
        sorted_vals = sorted(values)
        count = len(values)
        mean = calculate_mean(values)
        std = calculate_std(values)
        min_val = min(values)
        max_val = max(values)
        q25 = calculate_percentile(sorted_vals, 25)
        q50 = calculate_percentile(sorted_vals, 50)  # median
        q75 = calculate_percentile(sorted_vals, 75)
        
        # Store results
        numeric_stats_python[col] = {
            'count': count,
            'mean': mean,
            'std': std,
            'min': min_val,
            '25%': q25,
            '50%': q50,
            '75%': q75,
            'max': max_val
        }
        
        # Print formatted row with better number formatting
        print(f"{col:<45} {count:<8} {mean:<10.3f} {std:<10.3f} {min_val:<10.3f} {q25:<10.3f} {q50:<10.3f} {q75:<10.3f} {max_val:<12.3f}")
    else:
        print(f"{col:<45} {'No data'}")

print("\nAdditional Numeric Statistics (sum, variance):")
print(f"{'Column':<45} {'sum':<15} {'variance':<15}")
print("-" * 80)

for col in numeric_columns:
    values = get_numeric_values(facebook_data, col)
    if values:
        total_sum = sum(values)
        mean = calculate_mean(values)
        variance = sum((x - mean) ** 2 for x in values) / len(values) if values else 0
        
        # Format large numbers better
        if total_sum >= 1000000:
            sum_formatted = f"{total_sum:,.0f}"
        else:
            sum_formatted = f"{total_sum:.3f}"
            
        print(f"{col:<45} {sum_formatted:<15} {variance:<15.3f}")

# CATEGORICAL STATISTICS
print("\n" + "=" * 60)
print("CATEGORICAL COLUMNS STATISTICS")
print("=" * 60)

categorical_stats_python = {}
for col in categorical_columns:
    print(f"\n--- {col} ---")
    
    # Extract non-null values
    values = []
    for row in facebook_data:
        val = row.get(col)
        if val is not None and val != '':
            values.append(str(val))
    
    count = len(values)
    unique_values = list(set(values))
    unique_count = len(unique_values)
    
    # Most frequent value
    if values:
        counter = Counter(values)
        mode_item = counter.most_common(1)[0]
        mode_value = mode_item[0]
        mode_count = mode_item[1]
    else:
        mode_value = 'N/A'
        mode_count = 0
    
    print(f"Count: {count}")
    print(f"Unique Values: {unique_count}")
    print(f"Most Frequent: '{mode_value}' (appears {mode_count} times)")
    
    # Top 5 values
    print("Top 5 Values:")
    if values:
        counter = Counter(values)
        top_5 = counter.most_common(5)
        for i, (val, count_val) in enumerate(top_5, 1):
            print(f"  {i}. '{val}': {count_val}")
    
    # Store results
    categorical_stats_python[col] = {
        'count': count,
        'unique': unique_count,
        'mode': mode_value,
        'mode_count': mode_count
    }

# End timing
end_time = time.time()
execution_time = end_time - start_time

print("\n" + "=" * 60)
print("EXECUTION SUMMARY")
print("=" * 60)
print(f"Pure Python Part 1 execution time: {execution_time:.4f} seconds")
print(f"Dataset shape: ({len(facebook_data)}, {len(facebook_data[0].keys()) if facebook_data else 0})")
print(f"Numeric columns analyzed: {len(numeric_columns)}")
print(f"Categorical columns analyzed: {len(categorical_columns)}")

# Save results for comparison
python_results_part1 = {
    'numeric_stats': numeric_stats_python,
    'categorical_stats': categorical_stats_python,
    'execution_time': execution_time,
    'dataset_shape': (len(facebook_data), len(facebook_data[0].keys()) if facebook_data else 0)
}

print(f"\nResults saved in 'python_results_part1' variable for comparison")
```

```python
import time
import math
from collections import defaultdict, Counter

# Script 1 Part 2: Pure Python Groupby Analysis for Facebook Dataset
print("=== SCRIPT 1 PART 2: PURE PYTHON GROUPBY ANALYSIS ===")

# Start timing
start_time = time.time()

# Use the same data from Part 1
facebook_data = facebook_posts_clean.to_dict('records')

# Key columns for groupby analysis (same as Pandas script)
engagement_columns = ['Likes', 'Comments', 'Shares', 'Love', 'Wow', 'Haha', 'Sad', 'Angry', 'Care']
view_columns = ['Post Views', 'Total Views', 'Total Interactions']
topic_columns = [
    'covid_topic_illuminating', 'economy_topic_illuminating', 'education_topic_illuminating',
    'environment_topic_illuminating', 'foreign_policy_topic_illuminating', 'governance_topic_illuminating',
    'health_topic_illuminating', 'immigration_topic_illuminating', 'safety_topic_illuminating'
]

# HELPER FUNCTIONS FOR GROUPBY
def get_numeric_value(value):
    """Convert value to float, return None if not possible"""
    if value is None or (isinstance(value, float) and math.isnan(value)):
        return None
    try:
        return float(value)
    except (ValueError, TypeError):
        return None

def calculate_group_stats(values):
    """Calculate count, sum, mean, max for a list of values"""
    clean_values = [v for v in values if v is not None]
    
    if not clean_values:
        return {'count': 0, 'sum': 0, 'mean': 0, 'max': 0}
    
    return {
        'count': len(clean_values),
        'sum': sum(clean_values),
        'mean': sum(clean_values) / len(clean_values),
        'max': max(clean_values)
    }

print("=" * 70)
print("1. GROUPED BY FACEBOOK_ID (Page Analysis)")
print("=" * 70)

# Group data by Facebook_Id
facebook_groups = defaultdict(list)
for row in facebook_data:
    facebook_id = row.get('Facebook_Id')
    if facebook_id:
        facebook_groups[facebook_id].append(row)

print(f"\nFound {len(facebook_groups)} unique Facebook pages")

# Calculate page activity summary
print("\nPage Activity Summary:")
print(f"{'Facebook_Id':<70} {'Posts':<8} {'Likes_Sum':<12} {'Likes_Mean':<12} {'Comments_Sum':<12} {'Comments_Mean':<12}")
print("-" * 130)

page_stats = {}
for facebook_id, posts in facebook_groups.items():
    # Post count
    post_count = len(posts)
    
    # Likes stats
    likes_values = [get_numeric_value(post.get('Likes')) for post in posts]
    likes_stats = calculate_group_stats(likes_values)
    
    # Comments stats
    comments_values = [get_numeric_value(post.get('Comments')) for post in posts]
    comments_stats = calculate_group_stats(comments_values)
    
    # Shares stats
    shares_values = [get_numeric_value(post.get('Shares')) for post in posts]
    shares_stats = calculate_group_stats(shares_values)
    
    # Total Interactions stats
    interactions_values = [get_numeric_value(post.get('Total Interactions')) for post in posts]
    interactions_stats = calculate_group_stats(interactions_values)
    
    # Store results
    page_stats[facebook_id] = {
        'post_count': post_count,
        'likes': likes_stats,
        'comments': comments_stats,
        'shares': shares_stats,
        'interactions': interactions_stats
    }
    
    # Print first 10 pages
    if len(page_stats) <= 10:
        print(f"{facebook_id:<70} {post_count:<8} {likes_stats['sum']:<12.0f} {likes_stats['mean']:<12.3f} {comments_stats['sum']:<12.0f} {comments_stats['mean']:<12.3f}")

# Top 5 pages by different metrics
print("\nTop 5 Pages by Total Likes:")
likes_ranking = sorted(page_stats.items(), key=lambda x: x[1]['likes']['sum'], reverse=True)[:5]
for i, (page_id, stats) in enumerate(likes_ranking, 1):
    print(f"{i}. {page_id[:50]}... : {stats['likes']['sum']:.0f}")

print("\nTop 5 Pages by Total Comments:")
comments_ranking = sorted(page_stats.items(), key=lambda x: x[1]['comments']['sum'], reverse=True)[:5]
for i, (page_id, stats) in enumerate(comments_ranking, 1):
    print(f"{i}. {page_id[:50]}... : {stats['comments']['sum']:.0f}")

print("\nTop 5 Pages by Post Count:")
posts_ranking = sorted(page_stats.items(), key=lambda x: x[1]['post_count'], reverse=True)[:5]
for i, (page_id, stats) in enumerate(posts_ranking, 1):
    print(f"{i}. {page_id[:50]}... : {stats['post_count']}")

# Topic coverage by page
print("\nTopic Coverage by Page (Top 5 pages for each topic):")
for topic in topic_columns[:3]:  # Show first 3 topics
    print(f"\n{topic}:")
    topic_by_page = {}
    
    for facebook_id, posts in facebook_groups.items():
        topic_values = [get_numeric_value(post.get(topic)) for post in posts]
        topic_sum = sum(v for v in topic_values if v is not None)
        topic_by_page[facebook_id] = topic_sum
    
    # Sort and show top 5
    topic_ranking = sorted(topic_by_page.items(), key=lambda x: x[1], reverse=True)[:5]
    for i, (page_id, topic_sum) in enumerate(topic_ranking, 1):
        print(f"  {i}. {page_id[:50]}... : {topic_sum}")

print("\n" + "=" * 70)
print("2. GROUPED BY FACEBOOK_ID + PAGE CATEGORY")
print("=" * 70)

# Group by Facebook_Id AND Page Category
page_category_groups = defaultdict(list)
for row in facebook_data:
    facebook_id = row.get('Facebook_Id', '')
    page_category = row.get('Page Category', 'Unknown')
    
    # Handle NaN values
    if page_category is None or (isinstance(page_category, float) and math.isnan(page_category)):
        page_category = 'Unknown'
    
    key = (facebook_id, str(page_category))
    page_category_groups[key].append(row)

print(f"\nFound {len(page_category_groups)} unique Facebook page + category combinations")

# Calculate page + category stats
print("\nPage + Category Activity Summary (first 15):")
print(f"{'Facebook_Id':<50} {'Category':<20} {'Posts':<8} {'Likes_Sum':<12} {'Likes_Mean':<12}")
print("-" * 110)

page_cat_stats = {}
for (facebook_id, category), posts in list(page_category_groups.items())[:15]:
    post_count = len(posts)
    
    likes_values = [get_numeric_value(post.get('Likes')) for post in posts]
    likes_stats = calculate_group_stats(likes_values)
    
    page_cat_stats[(facebook_id, category)] = {
        'post_count': post_count,
        'likes': likes_stats
    }
    
    print(f"{facebook_id[:50]:<50} {category:<20} {post_count:<8} {likes_stats['sum']:<12.0f} {likes_stats['mean']:<12.3f}")

# Average engagement by Page Category
print("\nAverage Engagement by Page Category:")
category_groups = defaultdict(list)
for row in facebook_data:
    category = row.get('Page Category', 'Unknown')
    if category is None or (isinstance(category, float) and math.isnan(category)):
        category = 'Unknown'
    category_groups[str(category)].append(row)

print(f"{'Category':<25} {'Count':<8} {'Likes_Mean':<12} {'Comments_Mean':<15} {'Shares_Mean':<12}")
print("-" * 80)

category_stats = {}
for category, posts in category_groups.items():
    likes_values = [get_numeric_value(post.get('Likes')) for post in posts]
    comments_values = [get_numeric_value(post.get('Comments')) for post in posts]
    shares_values = [get_numeric_value(post.get('Shares')) for post in posts]
    
    likes_stats = calculate_group_stats(likes_values)
    comments_stats = calculate_group_stats(comments_values)
    shares_stats = calculate_group_stats(shares_values)
    
    category_stats[category] = {
        'count': len(posts),
        'likes_mean': likes_stats['mean'],
        'comments_mean': comments_stats['mean'],
        'shares_mean': shares_stats['mean']
    }
    
    print(f"{category:<25} {len(posts):<8} {likes_stats['mean']:<12.3f} {comments_stats['mean']:<15.3f} {shares_stats['mean']:<12.3f}")

# Post Type Distribution by Page Category
print("\nPost Type Distribution by Page Category:")
type_category_crosstab = defaultdict(lambda: defaultdict(int))

for row in facebook_data:
    category = row.get('Page Category', 'Unknown')
    post_type = row.get('Type', 'Unknown')
    
    if category is None or (isinstance(category, float) and math.isnan(category)):
        category = 'Unknown'
    if post_type is None or (isinstance(post_type, float) and math.isnan(post_type)):
        post_type = 'Unknown'
    
    type_category_crosstab[str(category)][str(post_type)] += 1

# Get all unique post types and categories
categories = sorted(type_category_crosstab.keys())
all_types = set()
for cat_types in type_category_crosstab.values():
    all_types.update(cat_types.keys())
all_types = sorted(all_types)

# Display as a cleaner table
print(f"{'Category':<25}", end="")
for post_type in all_types:
    print(f"{post_type:<12}", end="")
print(" | Total")
print("-" * (25 + 12 * len(all_types) + 8))

for category in categories:
    print(f"{category:<25}", end="")
    row_total = 0
    for post_type in all_types:
        count = type_category_crosstab[category].get(post_type, 0)
        print(f"{count:<12}", end="")
        row_total += count
    print(f" | {row_total}")

# Column totals
print(f"{'TOTAL':<25}", end="")
for post_type in all_types:
    col_total = sum(type_category_crosstab[cat].get(post_type, 0) for cat in categories)
    print(f"{col_total:<12}", end="")
grand_total = sum(sum(cat_types.values()) for cat_types in type_category_crosstab.values())
print(f" | {grand_total}")

# End timing
end_time = time.time()
execution_time = end_time - start_time

print("\n" + "=" * 70)
print("EXECUTION SUMMARY")
print("=" * 70)
print(f"Pure Python Part 2 execution time: {execution_time:.3f} seconds")
print(f"Unique Facebook_Id values: {len(facebook_groups)}")
print(f"Unique Page Categories: {len(category_groups)}")
print(f"Total page-category combinations: {len(page_category_groups)}")

# Save results for comparison
python_results_part2 = {
    'page_stats': page_stats,
    'category_stats': category_stats,
    'execution_time': execution_time,
    'facebook_groups_count': len(facebook_groups),
    'category_groups_count': len(category_groups)
}

print(f"\nResults saved in 'python_results_part2' variable for comparison")
```
